```python
import pandas as pd
print(pd.__version__)
df = pd.read_csv('data/scores.csv')
df = df.head(30)
df.index = range(100,3100,100)
df.head(3)
```

# 마지막에 행 추가하고 인덱스 다시 지정하기
* df.append(추가할데이터, ignore_index=True)
* 추가할데이터 -> 딕셔너리 형태로 전달

```python
new_value = {'name':31,'kor':8,'eng':9,'math':10}
# pandas 2.0.0 버전 이후부터 append() 메서드가 완전히 제거
# df.append(new_value,ignore_index=True)
```

# 인덱스 지정하여 추가/수정하기
데이터프레임.loc[인덱스] = 추가/수정데이터

```python
# 인덱스 31 추가
df.loc[31] = [31,8,9,10]
df
```

```python
# 인덱스 35 추가
df.loc[35] = [35,9,10,2]
df
```

```python
df.tail()
```

```python
# 인덱스 30
df.loc[3000] = [30,4,8,3]
```

```python
df.tail()
```

# 행 삭제하기
* df.drop(index=[삭제할 인덱스 리스트],inplace=True)

```python
# 3000,31,35 삭제
df.drop(index=[3000,31,35])
```

```python
df.tail()
```

```python
df.drop(index=[3000,31,35],inplace=True)
```

```python
df.tail()
```

# 인덱스 변경하기
## 전체 인덱스명 변경하기
* 전체/일부 인덱스 변경가능

* 전체 <br>
df.index = 인덱스 리스트 <br>
인덱스명리스트의 항목 수는 인덱스 수와 동일해야한다.

* 일부 <br>
df.rename(index={현재 인덱스명:바꿀 인덱스명,...})

```python
# range(100,3100,100)
df.index = range(10,300,10)
df
```

## 특정 인덱스명 변경하기

```python
# 100 -> 'a', 200 -> 'b'
df.rename(index={100:'a',200:'b'},inplace=True)
```
