```python
import pandas as pd
df = pd.read_csv('data/scores.csv')
df.head()
```

# 열 추가/수정하기
* 데이터프레임[컬럼] = 추가할 데이터
* 데이터프레임[컬럼] = 수정할 데이터
컬럼이 존재하면 추가, 존재하지 않으면 수정된다.

## 열 추가하기

```python
# 학생번호 추가하기(1부터 시작하여 1씩 증가)
df['no'] = range(1,len(df)+1)
```

```python
df
```

```python
# sum 추가
df['sum'] = df['kor'] + df['eng'] + df['math']
```

```python
df
```

## 열 수정하기

```python
# 학생 번호 수정하기(100부터 시작해서 1씩 증가)
```

```python
df['no'] = df['no'] + 99
```

```python
df
```

# 열 삭제하기
* 데이터프레임.drop(columns=삭제할컬럼리스트,inplace=True)
* 존재하지 않는 컬럼명을 사용하면 에러 발생

```python
# no,sum 컬럼 삭제하기
df.drop(columns=['no','sum'])
```

```python
df
```

```python
# drop을 데이터 프레임에 적용하기 위해선 inplace=Ture 속성 추가 필요
df.drop(columns=['no','sum'], inplace=True)
df
```

# 컬럼명 바꾸기

## 컬럼명 한번에 바꾸기
* 데이터프레임.columns = 컬럼명리스트 <br>
컬럼명리스트의 항목 수는 컬럼수와 동일해야한다

```python
df.head(1)
```

```python
# '이름','국어','영어','수학'
df.columns = ['이름','국어','영어','수학']
```

```python
df.head()
```

```python
# 컬럼수 != 리스트 항목 수 => error
df.columns = ['이름','국어','영어',]
```

```python
# 이름 -> 성명
df.rename(columns = {'이름':'성명'})
```
