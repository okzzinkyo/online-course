```python
import pandas as pd
df = pd.read_csv('data/titanic.csv')
df
```

(1) Survived가 1이고, PClass가 3이고, Sex가 male이고, 연령이 20대인 Name, Age를 추출하여 새로운 데이터프레임 df1으로 생성하시오.

```python
df1 = df[(df['Survived']==1) & (df['Pclass']==3) & (df['Sex']=='male') & (df['Age']//10==2)][['Name','Age']]
df1
```

(2) (1)에서 만든 df1에서 행번호가 행번호가 5,6인 데이터를 추출하시오.

```python
df1.iloc[[5,6]]
```
