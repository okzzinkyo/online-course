```python
import pandas as pd
df = pd.read_csv('data/temp.csv', encoding='cp949')
df
```

```python
import matplotlib.pyplot as plt

x = df['날짜']
y1 = df['평균기온']
y2 = df['최저기온']
y3 = df['최고기온']

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.grid()
plt.title('2022, Temperatures')
plt.xlabel('date')
plt.xlabel('temperatures')

plt.show()
```
