##  조건에 따라 데이터 추출하기
### 조건의 결과에 따른 불린인덱스 투툴

```python
# kor 점수가 10점인 데이터 불린 인덱스
df['kor'] == 10
```

### 불린인덱스로 데이터 추출하기

```python
df[df['kor'] == 10]
```

### 여러조건
* 논리연산자('&','|','~','^')를 사용할 때는 각 조건을 ()로 감싼다.

```python
# 한 과목이라도 10을 받은 학생 추출
df[(df.kor == 100) | (df.eng == 10) | (df.math == 10)]
```

```python
# kor의 값의 60~90 인 학생의 name, kor 추출
df[(df['kor'] >= 6)&(df['kor'] <= 9)][['name','kor']]
```

### 특정 값을 가진 데이터만 추출
* 컬럼.isin(값리스트)

```python
# 이름이 5인 데이터 추출
df[df['name'].isin([5])]
```

```python
# 이름이 5, 9인 데이터추출
df[df['name'].isin([5,9])]
```

```python
# kor이 5,10인 데이터 추출
df[df['kor'].isin([1,5])]
```

### null 여부에 따른 데이터 추출
* isnull() 컬럼의 값이 null인 데이터
* isnotnull() 컬럼의 값이 null이 아닌 데이터

```python
# kor이 null인 데이터 추출
df[df['kor'].isnull()]
```

```python
# kor이 null이 아닌 데이터 추출
df[df['kor'].notnull()]
```
