```python
# 6.남아출생이 가장 많은 top5를 구하시오(년도와 남아수)
df.sort_values(by='남아수',ascending=False).iloc[:5]
```

```python
# [풀이]
df['남아수'].nlargest()
```

```python
# 7.남아율(%) 컬럼을 추가하고 각년도별 남아수가 차지하는 비율을 데이터로 보여주시오.
# 컬럼 추가: df['남아율(%)']
# 연도별 남아수,여아수 합계: df.sum(axis=1)
df['남아율']= round((df['남아수']/df.sum(axis=1))*100,2)
df
```

```python
df.drop(columns=['남아율(%)'],inplace=True)
```

```python
# 8.2000년도 이후 데이터에 대해 출생량 컬럼을 추가하고, 남아수가 1800000이상이면 '많음' 아니면 '적음'이라고 보여주시오.
# df.loc[2000:]['출생량']
def fn(num):
    if num >= 1800000:
        return '많음'
    else:
        return '적음'
```

```python
df1 = df.loc[2000:][['남아수']]
df1['출생량'] = df1['남아수'].apply(fn)
df1
```

```python
# [풀이]
df2000 = df.loc[2000:][['남아수']]
df2000['출생량'] = df2000['남아수'].apply(lambda v: '많음' if v>=1800000 else '적음')
df2000
```
