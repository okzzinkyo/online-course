```python
# 1. 서울에서 일어난 요일별 사망자수, 사상자수를 출력하시오.
df1 = df[df['발생지시도'] == '서울'].groupby(by='요일')[['사망자수','사상자수']]
df1.sum()
```

```python
# [풀이]
seoulDF = df[df['발생지시도'] == '서울'][['요일','사망자수','사상자수']
seoulDF.groupby('요일')
df1.sum().loc[['월','화','수','목','금','토','일']]
```

```python
# 2. 서울 강북구 지역의 교통사고 위치를 지도로 표시하고 각 마커에는 법규위반내용을 표시하시오.
def fnMark(v):
    mk = folium.Marker(location=(v['위도'], v['경도']), popup=v['법규위반'])
    return mk
```

```python
df2 = df[(df['발생지시도'] == '서울') & (df['발생지시군구'] == '강북구')][['법규위반','위도','경도']]
map = folium.Map(location=(37.5012748,127.039625))
df2.apply(lambda v: fnMark(v).add_to(map), axis=1)
map
```

```python
# [풀이]
kangbuk =df[(df['발생지시도'] == '서울') & (df['발생지시군구'] == '강북구')][['법규위반','위도','경도']]
map = folium.Map(location=(kangbuk['위도'].mean(),kangbuk['경도'].mean()), zoom_start=14)
for i,s in kangbuk.iterrows():
    mk = folium.Marker(location=(s['위도'],s['경도']),popup=s['법규위반'])
    mk.add_to(map)

map.save('kangbuk.html')
```
