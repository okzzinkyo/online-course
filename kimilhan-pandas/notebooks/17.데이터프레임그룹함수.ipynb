```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family']="AppleGothic"
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_excel('../data/hotel.xlsx',index_col='Unnamed: 0')
df
```

```python
df.info()
```

```python
#등급별 가격의 합
g=df.groupby('grade')
print(g)

# groupby는 DataFrameGroupBy object를 반환
# sum,mean,max,min,agg,count 멤버함수가 있다.
g.sum()[['price']]
```

```python
# 등급별 가격의 평균
g.mean()[['price']]
```

```python
# 등급별 가격의 최댓값
g.max()[['price']]
```

```python
# 등급별 가격의 최솟값
g.min()[['price']]
```

```python
# 등급별 계수
# df['grade'].value_counts()
g.count()[['price']]
```

```python
# 등급별 평균 + 합
# 함수 또는 리스트로 적용할 함수를 할당해준다.
df1=g.agg(['sum','mean'])['price']
df1.columns = ['가격의합','가격의평균']
df1
```

```python
def fn(v):
    print('v=',v)
    print('=====')
    return v.sum()
```

```python
g.agg(fn)
```

```python
def fn1(v):
    # return v.mean()
    if v.mean()>8000:
        return "높다"
    else:
        return "낮다"
```

```python
g.agg(fn1)
```

```python
# g.agg(lambda v:v.mean())
g.agg(lambda v: "높다" if v.mean()>8000 else "낮다")
```

```python
dfGrade1 = pd.read_csv('../data/grade.csv')
dfGrade1
```

```python
dfGrade1.info()
```

```python
# 학년별 기말, 중간고사 평균을 구하시오
g2 = dfGrade1.groupby('학년')[['중간', '기말']]
g2.mean()

# pandas 버전에 따라 mean()이 object형 컬럼과 혼합된 DataFrame에서
# 예상치 못한 동작을 하며 에러를 내는 사례가 있습니다
# pandas 1.5 이후 버전에서는 groupby.mean()의 내부 동작 방식이 바뀌면서
# object 컬럼이 포함된 DataFrame에서 TypeError가 발생할 수 있다.
```

```python
# 학년별 과목별
g2 = dfGrade1.groupby(['학년','과목'])[['중간', '기말']]
g2.mean()
```
