```python
# 종목코드, 날짜
# hoo는 2017년 이후 비공식 API 사용에 대해 자주 차단하거나 URL 포맷을 변경해왔다. 
# pandas_datareader는 Yahoo Finance 지원을 공식적으로 중단하지는 않았지만,
# 종종 이렇게 작동하지 않는 경우가 있다.
# 방법 1) yfinance 사용
# 방법 2) pandas_datareader 최신 버전 확인
# sdf = data.get_data_yahoo('005930.KS', '2017-01-01')
# sdf
import yfinance as yf
import datetime
today = datetime.datetime.today().strftime('%Y-%m-%d')
sdf = yf.download('005930.KS', start='2017-01-01', end=today)
sdf
```

```python
# sdf.index가 datetype 형이기 때문에 year, month, day와 같은 속성을 쓸 수 있다.
sdf['year'] = sdf.index.year
sdf['month'] = sdf.index.month
sdf
```

```python
# g=sdf.groupby('year')
# 이렇게 컬럼을 만들지 않아도 바로 구할 수 있다.
g=sdf.groupby(sdf.index.year)
g.mean()
```

```python
gM=sdf.groupby(sdf.index.month)
gM.mean()
```

```python
gYM=sdf.groupby([sdf.index.year,sdf.index.month])
gYM.mean()
```

```python
# resample 이라는 함수를 통해 주간별, 분기별 등 다양한 집계를 할 수 있다.
# resample rule
# https://pandas.pydata.org/docs/user_guide/timeseries.html#time-date-components

# 연도별 직계
gRes = sdf.resample(rule='A')
gRes.mean()
```

```python
# 주간별(월~일)
gRes = sdf.resample(rule='w')
gRes.mean()
```

```python
# 분기별 직계
gRes = sdf.resample(rule='Q')
gRes.mean()
```
