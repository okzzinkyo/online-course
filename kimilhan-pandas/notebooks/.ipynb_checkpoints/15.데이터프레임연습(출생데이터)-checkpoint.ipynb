```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family']="AppleGothic"
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
# 첫번째 데이터를 컬럼명으로 부여하기 때문에 header=None 속성값 필요
# index_col=index로 설정할 컬럼
df = pd.read_csv('../data/births.txt', header=None, index_col=0)
df.columns = ['남아수', '여아수']
df.index.name = '연도'
df
```

```python
# [문제]
# 1.남아수 총 합을 구하시오.
# 2.여아수 평균을 구하시오.
# 3.남아수가 가장 많은 년도와 남아수를 구하시오.
# 4.200년도 이후 데이터를 구하시오.
# 5.1995~2000년도 데이터를 구하시오.
```

```python
# 1.남아수 총 합을 구하시오.
# df.loc[:,'남아수'].sum()
# df.sum()['남아수']
df['남아수'].sum()
```

```python
# 2.여아수 평균을 구하시오.
# df['여아수'].mean()
round(df['여아수'].mean(),2)
```

```python
# 3.남아수가 가장 많은 년도와 남아수를 구하시오.
# df[df['남아수'] == df['남아수'].max()][['남아수']]
mx = df['남아수'].max()
df.query(f'남아수=={mx}')[['남아수']]
```

```python
# 4.2000년도 이후 데이터를 구하시오.
# df[df.index >= 2000]
df.loc[2000:]
```

```python
# 5.1995~2000년도 데이터를 구하시오.
# df[(df.index >= 1955) & (df.index <= 2000)]
df.loc[1955:2000] # 부여된 index로 끝에 까지 포함
```
